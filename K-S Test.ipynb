```python
# KS -Test 
# KS test uses hypothesis testing
# To know whether two random variables have same distributions or not
```

```python
#Imporing the library
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Gaussian distribution
x = stats.norm.rvs(size=10);
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()
```

```python
stats.kstest(x, 'norm')
```

```python
#Looking at the p-value we can conclude X and Y have same distribution
```

```python
# Uniform Distribution
y = np.random.uniform(0,1,10000);
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```

```python
#Looking at the p-value we can conclude X and Y have different distribution
```

```python
# Pareto Distribution
a, m = 3., 2. 
x = (np.random.pareto(a, 1000) + 1) * m
sns.kdeplot(np.array(x), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```

```python
#Looking at the p-value we can conclude X and Y have different distribution
```

```python
# Weibull distribution
s =np.random.weibull(a, 1000)
sns.kdeplot(np.array(s), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```

```python
#Looking at the p-value we can conclude X and Y have different distribution
```
