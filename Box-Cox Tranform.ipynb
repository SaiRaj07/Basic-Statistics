```python
# Box-Cox Tranform
# Takes any distribution X and transforms to normal distribution
```

```python
#Importing the library
import numpy as np 
import pylab 
import scipy.stats as stats
```

```python
# Q-Q Plots (log Gamma distribution)
X = stats.loggamma.rvs(5, size=500) + 5
stats.probplot(X, dist="norm", plot=pylab)
pylab.show()
```

```python
# Here we can observe that X and Y have different distributions
```

```python
# Box-Cox Transform (Log-Gamma to Normal Dsitribution)
xt, _ = stats.boxcox(X)
stats.probplot(xt, dist="norm", plot=pylab)
pylab.show()
```

```python
# Thus we can conclude that box-cox transform is succesfull
```

```python
# Q-Q Plots (Power Law distribution)
from scipy.stats import powerlaw
a = 1.66
mean, var, skew, kurt = powerlaw.stats(a, moments='mvsk')
x = np.linspace(powerlaw.ppf(0.01, a),powerlaw.ppf(0.99, a), 10000)
stats.probplot(X, dist="norm", plot=pylab)
pylab.show()
```

```python
# Box-Cox Transform (Power law to Normal Dsitribution)
xt, _ = stats.boxcox(x)
stats.probplot(xt, dist="norm", plot=pylab)
pylab.show()
```

```python
#Thus we can conclude that box-cox transform is unsuccessful
```

```python
# Thus BOX-COX Transform is not always sucessfull in converting any given  distribution to Normal Distribution
```
