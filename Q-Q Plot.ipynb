```python
# Q-Q PLOT
# 1) Whether random variable X and Y are from same distribution
# 2)Is X normally distributed ??
```

```python
#Importing the library
import numpy as np 
import pylab 
import scipy.stats as stats
```

```python
# Q-Q Plots
X = np.random.normal(loc =15, scale = 10, size=10000) 
stats.probplot(X, dist="norm", plot=pylab)
pylab.show()
```

```python
# If all the points lie on the line and do not deviate every much from the line then we can say X and Y have same distribution
```

```python
# Here X is normal distribution and Y is normal distribution 
X = np.random.normal(loc =5, scale =5, size=100) 
stats.probplot(X, dist="norm", plot=pylab)
pylab.show()
```

```python
# If the the number of datapoints is less, Q-Q plot fails.
# We can see the same from above graph. We need more data when using Q-Q Plot
```

```python
# Here X is uniform distribution and Y is normal distribution 
X = np.random.uniform(low=-1, high=1, size=10000) 
stats.probplot(X, dist="norm", plot=pylab)
pylab.show()
```

```python
#Since the points do not lie on the line and deviate we can conclude X and Y have different distributions
```

```python
# Here X is pareto distribution and Y is normal distribution 
a, m = 3., 2. 
X = (np.random.pareto(a, 1000) + 1) * m
stats.probplot(X, dist="norm", plot=pylab)
pylab.show()
```

```python
#Since the points do not lie on the line and deviate we can conclude X and Y have different distributions
```
